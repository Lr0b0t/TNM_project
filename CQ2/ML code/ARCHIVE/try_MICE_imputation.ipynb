```julia
using DataFrames
using CSV
using Impute
using Mice
```

```julia
# Set up file paths
baseDir = joinpath("..", "final files")
trainFile = joinpath(baseDir, "train_features_Q2.csv")
testFile  = joinpath(baseDir, "test_features_Q2.csv")
train_out = joinpath(baseDir, "train_features_Q2_MICECHECK.csv")
test_out  = joinpath(baseDir, "test_features_Q2_MICECHECK.csv")

# Load data
train_df = CSV.read(trainFile, DataFrame)
test_df = CSV.read(testFile, DataFrame)

```

```julia
# Identify numeric columns
numeric_cols_train = names(train_df, Number)
numeric_cols_test = names(test_df, Number)

# Convert NaN to missing in all numeric columns (train)
for col in numeric_cols_train
    coldata = train_df[!, col]
    if eltype(coldata) <: AbstractFloat
        train_df[!, col] = ifelse.(isnan.(coldata), missing, coldata)
    end
end
numeric_cols_train
train_df
# Convert NaN to missing in all numeric columns (test)

for col in numeric_cols_test
    coldata = test_df[!, col]
    if eltype(coldata) <: AbstractFloat
        test_df[!, col] = ifelse.(isnan.(coldata), missing, coldata)
    end
end

train_df
```
